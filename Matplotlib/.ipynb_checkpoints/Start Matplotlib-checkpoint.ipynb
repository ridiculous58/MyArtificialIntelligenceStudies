```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ' %matplotlib inline ' yazdıgımız grafikleri anaconda içersinde direk gosterilmeye olanak saglar
```

```python
%matplotlib inline 
```

```python
yasListesi = [10,20,60,60,40,50,60,70,75,80] 
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi) 
```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"y")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo'nun yaşa gore değişimi")
# plt.show()  # %matplotlib inline yazmak isek bu komutu yazıcaz
```

```python
numpyArray = np.linspace(0,10,20)
```

```python
numpyArray
```

```python
numpyArrayCube = numpyArray ** 3
```

```python
numpyArrayCube
```

```python
plt.plot(numpyArray,numpyArrayCube,"g*-") #g--,g*- ,g
```

```python
plt.subplot(1,2,1) 
plt.plot(numpyArray,numpyArrayCube,"r*-")
plt.subplot(1,2,2) 
plt.plot(numpyArrayCube,numpyArray,"g--")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.8,0.8,0.3,0.3])
figureAxes.plot(numpyArray,numpyArrayCube,"g")
figureAxes.set_xlabel("X ekseni Veri İsmi")
figureAxes.set_ylabel("Y ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlıgı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyArray,numpyArrayCube,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyArrayCube,numpyArray,"b*-")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Kucuk Grafik Başlık")
```

## SubPlots

```python
# Subplotlar tuple dondurur , ve hem figure hemde eksen olusupta gelir
```

```python
(benimFigure,benimEksen) = plt.subplots()
benimEksen.plot(numpyArray,numpyArrayCube,"g")
```

```python
(benimFigure,benimEksen) = plt.subplots(nrows=1,ncols=2)
type(benimEksen) ## numpy.ndarray
```

```python
(benimFigure,benimEksenler) = plt.subplots(nrows=1,ncols=2)

for benimEksen in benimEksenler:
    benimEksen.plot(numpyArray,numpyArrayCube,"g")
plt.tight_layout() #iki grafigin arasına boşluk koyar bunu yapmaz ise biraz grafikler iç içe girer 
#benimFigure başka editor kullanıyorsak sonucu yazdırmak için tekrar cagrılır
```

```python
yeniFigur = plt.figure()
#dpi=150
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyArray,numpyArray ** 2, label = "Numpy Dizisi ** 2")
yeniEksen.plot(numpyArray,numpyArray ** 3, label = "Numpy Dizisi ** 3")
yeniEksen.legend(loc=1)
```

```python
yeniFigur.savefig("benimfigur.png",dpi=200)
```
