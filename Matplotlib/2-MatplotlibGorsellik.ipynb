```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
numpyDizisi2
```

```python
(benimFigur,benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,color="#6eabe0",alpha=0.3) # opacity kavramı alpha ile yapılır 
benimEksen.plot(numpyDizisi2,numpyDizisi1,color="#7dc51d")
```

```python
(yeniFigur,yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 2 , color="blue",linewidth = 1.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 3 , color="yellow",linewidth = 1.0)

yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 4 , color="#10a28e",linestyle = "-")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 5 , color="#10a28e",linestyle = "--")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 6 , color="#10a28e",linestyle = ":")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 7 , color="#10a28e",linestyle = "-.")

#marker ozelliği veri noktalarına verdiğin ifadeyi yerleştirir
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 8 , color="black",linestyle = "-",marker="+",markersize=4,markerfacecolor="red")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 9 , color="black",linestyle = "-",marker="o",markersize=8,markerfacecolor="red")
```
